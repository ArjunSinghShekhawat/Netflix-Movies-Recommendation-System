```python
import ast
data['genres'] = data['genres'].map(lambda x:ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x:ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x:ast.literal_eval(x))
```

```python
def get_name(x):
    name_list = []
    
    for i in x:

        if i.get("name")=="Science Fiction":
           scifi = 'Sci-Fi'
           name_list.append(scifi)
        else:
           name_list.append(i.get('name'))
    if name_list == []:
        return np.nan
    else:
        return " ".join(name_list)
```

```python
data['genres'] = data['genres'].apply(lambda x:get_name(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
data['actor_1_name'] = data['cast'].map(lambda x:get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
data['actor_2_name'] = data['cast'].map(lambda x:get_actor3(x))
```

```python

def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
data['actor_3_name'] = data['cast'].map(lambda x:get_actor3(x))
```

```python
data.head(2)
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```
